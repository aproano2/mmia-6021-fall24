## Bigrams

```python
b = {}
for w1, w2 in zip(tokens, tokens[1:]):
    bigram = (w1, w2)
    b[bigram] = b.get(bigram, 0) + 1
```

```python
uniq_tokens = sorted(list(set(tokens)))
```

```python
uniq_tokens
```

```python
n_tokens = len(uniq_tokens)
```

```python
N = 0.000001*torch.ones((n_tokens,n_tokens))
```

```python
ttoi = {t:i for i,t in enumerate(uniq_tokens)}
itot = {i:t for t,i in ttoi.items()}
```

```python
for w1, w2 in zip(tokens, tokens[1:]):
    ix1 = ttoi[w1]
    ix2 = ttoi[w2]
    N[ix1,ix2] += 1
```

```python
n_tokens
```

```python
N[ttoi['a']].sum()
```

```python
P = N.float()
P = P / P.sum(1, keepdim=True)
```

```python
P[ttoi['abejas']].sum()
```

```python
P.sum(1, keepdim=True)
# n_tokens x 1
```

```python
P.sum(1)
# n_tokens -> 1 x n_tokens
```

```python
P[ttoi['a']][:100]
```

```python
g = torch.Generator().manual_seed(13)
# Pr[a|b]
word1 = 'al'
ix = ttoi[word1]
p = P[ix]
jx = torch.multinomial(p, num_samples=1, replacement=True, generator=g).item()
word2 = itot[jx]
' '.join([word1, word2])

```

```python
g = torch.Generator().manual_seed(67)
for i in range(10):
    word = "aureliano"
    ix = ttoi[word]
    sentence = [word]
    while True:
        p = P[ix]
        ix = torch.multinomial(p, num_samples=1, replacement=True, generator=g).item()
        word = itot[ix]
        sentence.append(word)
        if '.' in word:
            break
    print(i, ' '.join(sentence))
```

```python
g = torch.Generator().manual_seed(67)
p = 1.0 * torch.ones(n_tokens) / n_tokens
for i in range(10):
    word = "los"
    ix = ttoi[word]
    sentence = [word]
    while True:
        ix = torch.multinomial(p, num_samples=1, replacement=True, generator=g).item()
        word = itot[ix]
        sentence.append(word)
        if '.' in word:
            break
    print(i, ' '.join(sentence))
```

### Likelihood

```python
log_likelihood = 0
n = 0
for w1, w2 in zip(tokens, tokens[1:]):
    ix1 = ttoi[w1]
    ix2 = ttoi[w2]
    pr = P[ix1,ix2]
    log_likelihood += torch.log(pr)
    n += 1

print(f'{log_likelihood=}')
nll = -log_likelihood
print(f'{nll=}')
nnll = nll/n
print(f'{nnll=}')
```
